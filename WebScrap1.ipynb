```python
def preproc(dat):
    dat.columns=dat.columns.str.replace('\n','')
    dat.dropna(axis=0,how='all',inplace=True)
    dat.replace(['\n'],[''],inplace=True,regex=True)
    dat.replace(['\s\*$'],[''],inplace=True,regex=True)
    dat.replace(['^\s'],[''],inplace=True,regex=True)
    dat.replace([','],[''],inplace=True,regex=True)
    dat.replace([r'\b[a-zA-Z]\b'],np.nan,inplace=True,regex=True)
    dat=dat.apply(pd.to_numeric,errors='ignore')
    return dat
```

```python
df1=preproc(df1)
df1
```

```python
url="https://en.wikipedia.org/wiki/List_of_countries_by_total_health_expenditure_per_capita"
```

```python
res_obj1=urllib2.urlopen(url)
html_as_list=res_obj1.readlines()
res_obj1.close()
html_as_list
```

```python
lst1=[]
for lines in html_as_list:
    lines=str(lines,'UTF-8').strip()
    if re.search('table class',lines):
        lst1.append(lines)
```

```python
lst1
```

```python
table_title2=re.findall(r'"([^"]*)"',lst1[0])
table_title2=table_title2[0]
```

```python
res_obj2=urllib2.urlopen(url)
html_as_text=res_obj2.read()
HCE1=BeautifulSoup(html_as_text)
res_obj2.close()
```

```python
table=HCE1.find('table',attrs={'class':table_title2})
headers=[x.text for x in table.find_all('th')]
rows=[]
for row in table.find_all('tr'):  #iterate over all rows
    rows.append([data.text for data in row.find_all('td')])
rows
```

```python
df2=pd.DataFrame(rows,columns=headers)
df2
```

```python
df2=preproc(df2)
df2
```

```python
df2.to_csv('varunfile2.csv')
```
